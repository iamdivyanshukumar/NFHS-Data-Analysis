## Handling Missing Values with KNN Imputation

### Overview  
Some columns in the dataset have missing values, but less than 85%. Since the data is grouped by states and districts, **K-Nearest Neighbors (KNN) Imputation** is used to fill these gaps. This method finds the most similar entries and uses them to estimate missing values.
#

## Steps  
1. **Find columns with missing values**  
   - Select columns where missing values are **more than 0 but less than 85%**.  

2. **Apply KNN Imputer**  
   - Use **KNN with 5 neighbors** (`n_neighbors=5`).  
   - Fill missing values with the average of the closest matches.  

3. **Check the data**  
   - Compare missing values **before and after imputation** to ensure accuracy.  
mputation**.

---


```python
from sklearn.impute import KNNImputer


columns_to_impute = [col for col in df.columns if 0 < df[col].isnull().sum() < 601]

for col in columns_to_impute:
    print(f"Missing values in {col}:------------> {df[col].isnull().sum()}")

knn_imputer = KNNImputer(n_neighbors=5, weights="uniform")

df[columns_to_impute] = knn_imputer.fit_transform(df[columns_to_impute])

print("\n Missing values after KNN Imputation:")
print(df[columns_to_impute].isnull().sum())
```

# **Identifying Highly Correlated Columns**

### Step 1: Compute Correlation Matrix  
We calculate the **correlation matrix** for numerical columns, starting from column index **3**. Correlation values are taken in **absolute terms** to capture both positive and negative relationships.
#

## Step 2: Find Highly Correlated Columns  
- We use an **upper triangular mask** to avoid duplicate comparisons.  
- Columns with **correlation greater than 85%** are flagged.  
- If a column is highly correlated with another, it may be removed to reduce redunda#ncy.

## Step 3: Capture Correlation Pairs  
- A list of **highly correlated column pairs** is created.  
- This helps in identifying overlapping features that may affect a#nalysis.

## Results:  
Columns with **correlation above 85%** are identified, along with their correlation values.


```python
corr_matrix = df.iloc[:, 2:].corr()

mask = np.tril(np.ones(corr_matrix.shape), k=0).astype(bool)
corr_unstacked = corr_matrix.where(~mask)

sorted_corr = corr_unstacked.unstack().dropna().sort_values(ascending=False)

print("\n Top Highly Correlated Pairs (Correlation > 0.85)\n")
for (col1, col2), corr_value in sorted_corr.items():
    if corr_value > 0.85:  # Only show significant correlations
        print(f" {col1} \n----> {col2} \n --> {corr_value:.2f}")
```

# State-wise Data Aggregation  

## Objective  
To combine data at the state and UT level for better understanding.  

## Method  
- **Percentages and ratios** are averaged.  
- **Total counts** are summed.  
- **Text columns** keep the first entry.  

## Steps  
1. Choose the right method for each column.  
2. Group data by state/UT.  
3. Apply the methods.  
4. Reset the index.  

## Why It Matters  
- Keeps numbers accurate.  
- Avoids errors in calculations.  
- Makes state-wise comparison easy.  
ving data integrity!


```python
# Define aggregation functions
def custom_agg(column):
    if df[column].dtype in ['int64', 'float64']:  # Check if numeric column
        if "%" in column or "ratio" in column or "rate" in column or "per" in column.lower():
            return "mean"  # Take mean for percentage/ratio columns
        else:
            return "sum"  # Sum absolute count columns
    else:
        return "first"  # Keep the first entry for non-numeric columns

# Generate aggregation dictionary dynamically
agg_dict = {col: custom_agg(col) for col in df.columns if col != "State/UT"}

# Group by state and aggregate
df_states = df.groupby("State/UT").agg(agg_dict)

# Reset index if needed
df_states.reset_index(inplace=True)

df_states.drop(columns=['District Names'], inplace=True, errors='ignore')
# Display the result
df_states.head()
```

```python
#contert float dtype to int because per 1000 value is always in int data type
df_states[" Sex ratio of the total population (females per 1,000 males)"] = df_states[" Sex ratio of the total population (females per 1,000 males)"].astype(int)

df_states["Sex ratio at birth for children born in the last five years (females per 1,000 males)"] = df_states["Sex ratio at birth for children born in the last five years (females per 1,000 males)"].astype(int)
```

## Conclusion
High women's literacy does not guarantee better blood sugar health outcomes. Lifestyle, awareness, and access to healthcare are critical. States with both high literacy and high diabetes rates may benefit from focused health literacy and diabetes prevention campaigns.
